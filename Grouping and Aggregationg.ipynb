```python
# to find % of people from each country knows python

country_respodents = df['Country'].value_counts()
country_respodents 
```

```python
def count_python(x):    
    return x.str.contains('Python').sum()
country_uses_python = country_grp['LanguageWorkedWith'].apply(count_python)
country_uses_python
```

```python
python_df = pd.concat([country_respodents, country_uses_python], axis='columns')
python_df
```

```python
python_df.rename(columns={'Coutry': 'NumRespondents', 'LanguageWorkedWith':'NumKnowsPython'}, inplace=True)
python_df
```

```python
python_df['PctKnowsPython'] = python_df['NumKnowsPython']/python_df['Country'] * 100
```
